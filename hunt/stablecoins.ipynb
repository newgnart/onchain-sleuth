```python
import os, json
from pathlib import Path
import sys
import polars as pl
rootdir = Path.cwd().parent
sys.path.append(str(rootdir))

from onchaindata.utils.database import PostgresClient

client = PostgresClient.from_env()
```

## Some intresting transactions
Big amounts

```python
query = """
SELECT
    amount,
    token_address,
    from_address,
    to_address,
    transaction_hash
FROM staging.erc20_transfer
WHERE amount > 1e25 -- Adjust threshold
ORDER BY amount DESC
LIMIT 20;
"""
import pandas as pd
df = pd.read_sql(query, client.sqlalchemy_engine)
df
```

## ERC20 transfers count

```python
query = """
SELECT
      token_address,
      COUNT(*) as erc20_transfer_count
  FROM staging.erc20_transfer
  GROUP BY token_address
  ORDER BY erc20_transfer_count DESC
  LIMIT 20;
"""
import pandas as pd
df = pd.read_sql(query, client.sqlalchemy_engine)

with open(rootdir / 'scripts/stablecoins/contracts.json', 'r') as f:
      contracts = json.load(f)

  # Create address to name mapping (convert addresses to lowercase for matching)
address_to_name = {v['address'].lower(): k for k, v in contracts.items()}

# Add token name column to DataFrame
df['stablecoin'] = df['token_address'].str.lower().map(address_to_name)

# Fill NaN values with 'unknown' for tokens not in the mapping
df['stablecoin'] = df['stablecoin'].fillna('unknown')

df[['stablecoin', 'erc20_transfer_count']]
```
